```python
#Countour plots
import numpy as np
x = np.linspace(0,5,50)
y = np.linspace(0,5,50)
def func(x,y):
  return np.sin(x) ** 2 + np.cos(y)**2
X,Y = np.meshgrid(x,y)
Z=func(X,Y)
plt.contour(X,Y,Z,cmap='viridis')
```
